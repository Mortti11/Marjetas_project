```python
wind_df = pd.read_csv('open-meteo-62.39N25.62E139m (1).csv', skiprows=3)
```

```python
time_clean = (wind_df["time"].astype("string").str.strip().str.strip("'\"").replace(r"^(?:nan|null|none)?$", pd.NA, regex=True))


mask = time_clean.notna()
wind_df["Timestamp"] = pd.NaT
wind_df.loc[mask, "Timestamp"] = pd.to_datetime(time_clean[mask],format="ISO8601",errors="coerce")

print("Parsed:", mask.sum(), " | NaT:", wind_df["Timestamp"].isna().sum())
```

```python
wind_df.drop(columns=["time"], inplace=True)
```
